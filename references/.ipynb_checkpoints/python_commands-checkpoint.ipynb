`parse_dates`: bool or list of int or names or list of lists or dict, default False
               
   The behavior is as follows:

   bool (Boolean value). If True -> try parsing the index.
    
   list of int or names. e.g. If [1, 2, 3] -> try parsing columns 1, 2, 3 each as a separate date column.
    
   list of lists. e.g. If [[1, 3]] -> combine columns 1 and 3 and parse as a single date column.
    
   dict, e.g. {‘foo’ : [1, 3]} -> parse columns 1, 3 as date and call result ‘foo’

   If a column or index cannot be represented as an array of datetimes, say because of an unparsable value or a mixture of timezones, 
   the column or index will be returned unaltered as an object data type. For non-standard datetime parsing, use pd.to_datetime after 
   pd.read_csv. To parse an index or column with a mixture of timezones, specify date_parser to be a partially-applied 
   'pandas.to_datetime()' with utc=True. See Parsing a CSV with mixed timezones for more.

   Note: A fast-path exists for iso8601-formatted dates.

`infer_datetime_format`: bool, default False
     If True and parse_dates is enabled, pandas will attempt to infer the format of the datetime strings in the columns, 
     and if it can be inferred, switch to a faster method of parsing them. In some cases this can increase the parsing speed by 5-10x.
