# Sorting Pandas DataFrames

In this exercise you'll learn how to use sort Pandas DataFrames in ascending and descending order.

## Instructions

Perform the following:

1. Import the `pandas` libraries.

3. Read the CSV into a Pandas DataFrame.

4. Sort DataFrame in ascending order.

5. Sort DataFrame in descending order.

---

## 1. Import Libraries and Dependencies

```python
# Import libraries and dependencies
import pandas as pd
```

## 3. Read the CSV into a Pandas DataFrame

```python
# Read in the CSV as a DataFrame
df_pc = pd.read_csv("people_cleansed.csv")
df_pc.head()
```

## 4. Sort DataFrame in Ascending Order

```python
# Sort ascending (default)
df_pc.sort_values("Salary")
```

## 5. Sort DataFrame in Descending Order

```python
# Sort descending
df_pc.sort_values("Salary", ascending=False)
```
